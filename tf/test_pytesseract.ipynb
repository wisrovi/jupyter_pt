```python
from PIL import Image
import pandas as pd
import pytesseract
import os
from slugify import slugify
```

```python
df_train = pd.read_csv("/data/train.csv")
df_test = pd.read_csv("/data/test.csv")
df_valid = pd.read_csv("/data/valid.csv")
```

```python
def image_to_text_with_pytesseract(path_image: str) -> list:
    file = Image.open(os.path.join(path_image))
    gray = file.convert('L')

    text = pytesseract.image_to_data(gray, output_type='data.frame')
    # find confidence in text prediction
    text = text[text.conf != -1]
    text.head()
    lines = text.groupby(['page_num', 'block_num', 'par_num', 'line_num'])['text'] \
                     .apply(lambda x: ' '.join(list(x))).tolist()
    confs = text.groupby(['page_num', 'block_num', 'par_num', 'line_num'])['conf'].mean().tolist()

    line_conf = []
    for i in range(len(lines)):
        if lines[i].strip():
            text_ok = lines[i]
            #text_ok = slugify(text_ok)
            line_conf.append((text_ok, round(confs[i],3)))


    return line_conf

def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
```

```python
ACC = count = 0

len_dataset = df_train.shape[0]

for i in range(len_dataset):
    path_image, license_original = df_train["Path"][i], df_train["Content"][i]
    print(path_image)
    
    
    license_original_char = list(license_original)

    try:
        license_predict = image_to_text_with_pytesseract(path_image)
    except:
        print("fail: ",path_image.split(".")[1])
        continue    

    acc_tmp = 0
    for license_pred  in license_predict:
        license, conf = license_pred[0], license_pred[1]
        
        license_char = list(license)
        
        inters = intersection(license_original_char, license_char)
        
        maxim = len(license_original_char)
        detect = len(inters)
        acc_tmp = detect/maxim    
    
        if len(license_char)>0:
            print(path_image[:3] + "..." + path_image[-30:], "\t", license, "\t", license_original, "".join(inters), acc_tmp)
    
    count += 1
    ACC += acc_tmp
    
```
