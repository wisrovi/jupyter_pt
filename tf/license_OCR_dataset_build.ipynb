```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dossier = "/serverstorage"
chemins = []
contenus = []
```

```python
# folder matriculas

chemin_absolu_dossier = os.path.abspath(os.path.join(dossier, "matriculas"))

for fichier in os.listdir(chemin_absolu_dossier):
    if fichier.endswith(".txt"):
        chemin = os.path.join(chemin_absolu_dossier, fichier)
        chemins.append(chemin.split(".")[0]+".jpg")
        with open(chemin, "r") as f:
            contenu = f.read()
            contenus.append(contenu)
```

```python
chemin_absolu_dossier = os.path.abspath(os.path.join(dossier, "Matriculas_para_hacer"))

for fichier in os.listdir(chemin_absolu_dossier):
    
    chemin = os.path.join(chemin_absolu_dossier, fichier)
    contenu = fichier.split(".")[0]
    
    chemins.append(chemin)
    contenus.append(contenu)
```

```python
chemin_absolu_dossier = os.path.abspath(os.path.join(dossier, "Matriculas-Fran", "Nombre"))

for fichier in os.listdir(chemin_absolu_dossier):
    
    chemin = os.path.join(chemin_absolu_dossier, fichier)
    contenu = fichier.split(".")[0]
    
    chemins.append(chemin)
    contenus.append(contenu)
```

```python
chemin_absolu_dossier = os.path.abspath(os.path.join(dossier, "Matriculas-Fran", "Carpetas"))

for fichier in os.listdir(chemin_absolu_dossier):
    contenu = fichier
    contenu = contenu.replace("_", " ")
    
    for fichier2 in os.listdir(os.path.join(chemin_absolu_dossier, fichier)):
        chemin = os.path.join(chemin_absolu_dossier, fichier, fichier2)
        chemins.append(chemin)
        contenus.append(contenu)
```

```python
chemin_absolu_dossier = os.path.abspath(os.path.join(dossier, "Matriculas-Jesus", "RENOMBRADAS"))

for fichier in os.listdir(chemin_absolu_dossier):
    
    chemin = os.path.join(chemin_absolu_dossier, fichier)
    contenu = fichier.split(".")[0]
    
    chemins.append(chemin)
    contenus.append(contenu)
```

```python
chemin_absolu_dossier = os.path.abspath(os.path.join(dossier, "Matriculas-Jesus", "POR-CARPETAS"))

for fichier in os.listdir(chemin_absolu_dossier):
    if not fichier.find("."):
        contenu = fichier
        contenu = contenu.replace("_", " ")

        for fichier2 in os.listdir(os.path.join(chemin_absolu_dossier, fichier)):
            chemin = os.path.join(chemin_absolu_dossier, fichier, fichier2)
            chemins.append(chemin)
            contenus.append(contenu)
```

```python
chemins ,contenus
```

```python
X_train, X_test, y_train, y_test = train_test_split(chemins, contenus, test_size=0.2, random_state=0)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
len(X_train), len(y_train), len(X_test), len(y_test), len(X_valid), len(y_valid)
```

```python
df_train = pd.DataFrame({"Path": X_train, "Content": y_train})
df_test = pd.DataFrame({"Path": X_test, "Content": y_test})
df_valid = pd.DataFrame({"Path": X_valid, "Content": y_valid})
```

```python
df_train.to_csv(os.path.join("/data", "train.csv"), index=False)
df_test.to_csv(os.path.join("/data", "test.csv"), index=False)
df_valid.to_csv(os.path.join("/data", "valid.csv"), index=False)
```
